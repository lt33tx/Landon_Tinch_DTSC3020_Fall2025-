```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
penguins_df = sns.load_dataset("penguins")

# 1–3) Scatter plot with color by species
# Create a scatter plot comparing bill length and bill depth, colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=penguins_df,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",  # Color encoding by 'species'
    s=70,
    palette="viridis"
)
plt.title("Bill Length vs. Bill Depth of Penguins by Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.legend(title="Species")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

**Your interpretation (2–3 sentences):**

The scatter plot reveals three clearly distinct clusters, showing a strong separation of the penguin species based on their bill dimensions. Gentoo penguins tend to have the longest bills but the least depth, while Adelie penguins generally have the shortest but deepest bills. This visual pattern demonstrates that bill length and bill depth are excellent metrics for differentiating the three species.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (in case it wasn't loaded in the current session)
penguins_df = sns.load_dataset("penguins")

# TODO: Histogram with 10 bins
plt.figure(figsize=(7, 5))
sns.histplot(
    data=penguins_df,
    x="bill_length_mm",
    bins=10,  # Specify 10 bins
    kde=True  # Add a kernel density estimate line for smoothness
)
plt.title("Bill Length Distribution (10 Bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# TODO: Histogram with 30 bins
# TODO: Histogram with 30 bins
plt.figure(figsize=(7, 5))
sns.histplot(
    data=penguins_df,
    x="bill_length_mm",
    bins=30,  # Specify 30 bins
    kde=True
)
plt.title("Bill Length Distribution (30 Bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Frequency")
plt.show()
```

**Your comparison (1–2 sentences):**

The 10-bin histogram provides a generalized view, clearly suggesting a bimodal (two-peaked) distribution for bill length. In contrast, the 30-bin histogram uses finer detail that makes the distribution appear more jagged and slightly obscures the clear two-peaked pattern, though the underlying bimodal structure is still suggested by the two main regions of high frequency.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (or ensure it's loaded from previous steps)
penguins_df = sns.load_dataset("penguins")

# 5) Boxplot of body mass by species
# Create a boxplot of body_mass_g grouped by species
plt.figure(figsize=(8, 6))
sns.boxplot(
    data=penguins_df,
    x="species",
    y="body_mass_g",
    palette="pastel"
)
plt.title("Body Mass Distribution by Penguin Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()
```

**Your summary (1–2 sentences):**

The boxplot clearly shows significant differences in body mass between the species. The Gentoo penguins are the heaviest on average (highest median and distribution), while the Adelie species is the lightest.
