
## Part 1 – Initial Setup & Sample Data (0.5 point)

**Requirements:**

- Set a page title for the Streamlit app.
- Create an initial dataset with at least **5 sample books**.
- Each book should have the following fields:
  - `Title`  (string)
  - `Author` (string)
  - `Genre`  (string, e.g., "Fiction", "Non-Fiction", "Sci-Fi", etc.)
  - `Pages`  (integer)
  - `Read`   (boolean: `True` if read, `False` if unread)
- Store the data in a `pandas` DataFrame.

**Hint:**  
You can start from a list of dictionaries and then convert it to a DataFrame.

>  Write your code for **Part 1** in the next cell.


```python
%%writefile app.py

import streamlit as st
import pandas as pd

# set up the Streamlit page with a custom title
st.set_page_config(page_title="Pop Culture Reading List Tracker")

# starting list of books we'll load into the app
initial_books_data = [
    {"Title": "Harry Potter and the Sorcerer's Stone", "Author": "J.K. Rowling", "Genre": "Fantasy", "Pages": 320, "Read": True},
    {"Title": "The Hunger Games", "Author": "Suzanne Collins", "Genre": "Dystopian", "Pages": 374, "Read": True},
    {"Title": "Twilight", "Author": "Stephenie Meyer", "Genre": "Fantasy/Romance", "Pages": 498, "Read": False},
    {"Title": "The Girl with the Dragon Tattoo", "Author": "Stieg Larsson", "Genre": "Thriller", "Pages": 465, "Read": False},
    {"Title": "The Fault in Our Stars", "Author": "John Green", "Genre": "Young Adult", "Pages": 313, "Read": True},
    {"Title": "Fifty Shades of Grey", "Author": "E. L. James", "Genre": "Romance", "Pages": 514, "Read": False}
]

# only create the DataFrame once so it stays saved between interactions
if "books_df" not in st.session_state:
    st.session_state["books_df"] = pd.DataFrame(initial_books_data)
```


## Part 3 – Add New Book Form (1.0 point)

**Requirements:**

- Create a small form that allows the user to add a new book.
- The form must include inputs for:
  - `Title`  (text input)
  - `Author` (text input)
  - `Genre`  (selectbox or text input)
  - `Pages`  (number input, must be > 0)
  - `Read`   (checkbox)
- When the user submits the form:
  - **Validate** the inputs:
    - `Title` and `Author` cannot be empty.
    - `Pages` must be a positive number.
  - If validation passes:
    - Append the new book to the DataFrame.
    - Show a **success message**.
    - Update the displayed table.
  - If validation fails:
    - Show an **error message**.

**Hints:**
- You can use `st.form()` and `st.form_submit_button()` to create the form.
- To update the table, you must update the same DataFrame you used in Part 2.

>  Write your code for **Part 3** in the next cell.


```python
%%writefile -a app.py

# section where the user can add a brand-new book
st.subheader("Add a New Book")

# list of genres the user can pick from
GENRES = ["Fiction", "Non-Fiction", "Sci-Fi", "Fantasy", "Thriller", "Romance", "Dystopian", "Young Adult", "Memoir", "Other"]

# form so the user can input all the book details at once
with st.form("add_book_form"):
    col1, col2 = st.columns(2)  # split the form into two columns

    with col1:
        new_title = st.text_input("Title")  # user types the book title
        new_author = st.text_input("Author")  # user types the author name

    with col2:
        new_genre = st.selectbox("Genre", options=GENRES)  # dropdown for genre
        new_pages = st.number_input("Pages", min_value=1, step=1, value=1)  # number input for pages

    new_read = st.checkbox("Read", value=False)  # checkbox to mark if already read

    submitted = st.form_submit_button("Add Book to Library")  # submit button

    if submitted:
        # basic checks to make sure the user didn't leave anything important blank
        if not new_title.strip():
            st.error("Book Title cannot be empty.")
        elif not new_author.strip():
            st.error("Author cannot be empty.")
        elif new_pages < 1:
            st.error("Pages must be a positive number.")
        else:
            # create a small DataFrame for the new book entry
            new_book = pd.DataFrame([{
                "Title": new_title.strip(),
                "Author": new_author.strip(),
                "Genre": new_genre,
                "Pages": int(new_pages),
                "Read": new_read
            }])

            # add the new book to the main DataFrame stored in session state
            st.session_state["books_df"] = pd.concat([st.session_state["books_df"], new_book], ignore_index=True)

            # success message so the user knows it worked
            st.success(f"Successfully added '{new_title}' by {new_author}!")
```
