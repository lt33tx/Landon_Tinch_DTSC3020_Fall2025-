```python
import pandas as pd

df = pd.read_csv('sample_data/mtcars.csv')
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(df['mpg'], bins=8, edgecolor='black')
plt.title('Histogram of MPG Distribution')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df.boxplot(column='mpg', by='cyl', vert=True)
plt.title('Boxplot of MPG by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.suptitle('')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = df['wt']
y = df['mpg']

plt.figure()
plt.scatter(x, y)
plt.title('Scatter Plot of Weight vs. MPG')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.show()
```

```python
import matplotlib.pyplot as plt

counts = df['cyl'].value_counts().sort_index()

plt.figure()
plt.bar(counts.index.astype(str), counts.values)
plt.title('Bar Chart of Cylinder Counts')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt

df['hp'].plot(kind='kde', figsize=(8, 6))

plt.title('Kernel Density Estimation Plot of Horsepower Distribution')
plt.xlabel('Horsepower')
plt.ylabel('Density')
plt.show()
```
